```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
url = 'https://aves.ninjas.cl/api/birds'
```

```python
respuesta = requests.get(url)
print(respuesta)
```

```python
contenido = respuesta.text
DatosApi = json.loads(contenido)
df= pd.DataFrame.from_dict(DatosApi)
```

```python
dfname = json_normalize(df['name'])
dfname.head()
```

```python
df2 = pd.concat( [df,dfname], axis = 1)
df2.head()
```

```python
df3 = df2.drop(['name','sort'],axis=1)
df3.head()
```

```python
dfimages = json_normalize(df3['images'])
dfimages.head()
```

```python
df4 = pd.concat([df3,dfimages], axis= 1)
df4.head(3)
```

```python
df5 = df4.drop(['images'],axis =1)
df5.head(3)
```

```python
dflinks = json_normalize(df3['_links'])
dflinks.head(3)
```

```python
df6 = pd.concat([df5,dflinks],axis = 1)
df6.head(3)
```

```python
dffinal = df6.drop(['_links','parent'], axis=1)
dffinal.head(3)
```
